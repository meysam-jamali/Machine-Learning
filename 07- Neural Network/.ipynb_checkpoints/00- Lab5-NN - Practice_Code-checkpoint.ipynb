### Step 2

Step 2-1: Random Data Generation

```python
def generate_2d_cls_data(n, sigma_noise):
    # n_cls is the number of samples for each class
    # In this instance, only two classes are utitlized therefore, n is divided by 2
    # a half for each class
    n_cls = n // 2
    
    # polar coordinates (angles)
    # the use of angles (theta) is a design choice to create a dataset with
        # a circular or spiral pattern الگوی دایره ای یا مارپیچی
    # Generate random angles for the first class
    '''
        >>> theta = np.random.randn(10)
        >>> theta
        array([-0.32743906, -0.67775832,  0.43540703, -0.45671368, -0.3957138 ,
               -0.10545759,  0.70252693, -0.82650361, -0.83008829, -0.55297913])
        >>> theta = np.random.randn(10) * 2 * np.pi
        >>> theta
        array([ 6.96871455, -2.0655985 , -0.20193149, -4.77418229, -0.53852324,
               -3.91511193,  0.47810646, -0.67011709,  6.69588219,  6.92188582])
    '''
    # theta = 2*PI*Rand
    ''' 2 * np.pi: This scales the random numbers by 2 * np.pi, which is
             equivalent to a full circle in radians (360 degrees). Multiplying the
             random numbers by this factor ensures that they cover the entire range
             of possible angles in a circular coordinate system. '''
    # Generate n_cls number
    theta = np.random.randn(n_cls) * 2 * np.pi
    
    # Create the first class points based on polar coordinates (angles)
    ''' This stacks the x-coordinates and y-coordinates along a new axis 
            (axis 1). The result is a 2D array where each row represents a 
            point on the circle, and the first column contains x-coordinates,
            and the second column contains y-coordinates. Combines these 
            coordinates into a 2D array.'''

    cls1 = np.stack([np.cos(theta) * 2, np.sin(theta) * 2], axis=1)
    
    # Generate random points for the second class
    # Generate n_cls rows and two columns which multiplied by 0.5
    cls2 = np.random.randn(n_cls, 2) * 0.5

    # Add some noise to both classes
    # clsX.shape[0] the number of elements of each class
    cls1 += np.random.randn(cls1.shape[0], 2) * sigma_noise * 3
    cls2 += np.random.randn(cls2.shape[0], 2) * sigma_noise
    
    # Combine points from both classes to form the dataset
    # Classes are combined into X as the major dataset
    X = np.concatenate([cls1, cls2], axis=0)
    
    # Assign labels to the points (0 for class 1, 1 for class 2)
    # One class is labeled by zero, the other with one
    y = np.concatenate([np.zeros(n_cls), np.ones(n_cls)], axis=0)
    
    return X, y
```

Code Part 2 - separatingFLR

```python
# It's better to relocate this code 
def separatingFLR(data, labels, predictions, model):
    fig, ax = plt.subplots()
    
    # Set the number of grid points for creating a mesh
    num_grid_points = 100
    
    # Generate a grid of x and y values within the range of the data
    ''' >>> np.linspace(2.0,3.0, num=5)
        array([2.  , 2.25, 2.5 , 2.75, 3.  ]) '''
    # data[:, 0] means the first column
    # data[:, 1] means the second column
    # From minimum to maximum number num_grid_points number evenly spaced
    # values are generated
    xi = np.linspace(data[:, 0].min(), data[:, 0].max(), num_grid_points)
    yi = np.linspace(data[:, 1].min(), data[:, 1].max(), num_grid_points)
    # Mesh network
    X, Y = np.meshgrid(xi, yi)
    
    # Create a grid of feature points
    # Stack two matrixies xi and yi and combined them together at 1 axis column
    gdata = np.stack([xi, yi], 1)
    
    # Use the trained model to predict the labels for the grid points
    pred_grid = model.predict(np.stack([X.reshape(-1), Y.reshape(-1)], axis=1)).reshape(num_grid_points, num_grid_points)
    
    # Plot the contour lines of the decision boundary
    # Decision boundary = [0.5]
    ax.contour(xi, yi, pred_grid, 15, linewidths=2, colors='k', levels=[0.5])
    
    # Plot the data points
    ax.scatter(data[:,0], data[:,1], c=labels.ravel(), marker='o', s=100, zorder=10, alpha=0.8)
    
    # Set the x and y axis limits based on the range of the data
    ax.set_xlim(data[:,0].min(), data[:,0].max())
    ax.set_ylim(data[:,1].min(), data[:,1].max())
    
    return fig, ax
```

Code Part 3: Error Calculation

```python
def calc_err(predicted, observed):
    # Flatten the predicted and observed arrays
    ''' This function flattens the arrays, ensuring that they are
          1-dimensional. This is important for comparing individual elements 
            used here: predicted < 0.5'''
    predicted = predicted.ravel()
    observed = observed.ravel()
    
    # Create a copy of the predicted values to be used with different thresholds
    threshold_preds = predicted.copy()
    
    # Set predictions below 0.5 to 0, and predictions at or above 0.5 to 1
    threshold_preds[predicted < 0.5] = 0
    threshold_preds[predicted >= 0.5] = 1
    
    # Calculate the classification error as the mean of misclassified instances
    # Misclassified means the element couldn't be categorized in either classies
    return np.mean(threshold_preds != observed)
```

### Step 3: Executation

```python
# Step 1: Build the model 

# My Code
# Random Data Generation
Xtr, Ytr = generate_2d_cls_data(100, 0.01)
Xte, Yte = generate_2d_cls_data(20, 0.01)


# Build your first model by creating a Sequential object and then adding
	# 2 Dense layers:
'''
- `Sequential()`: This creates a linear stack of layers for the model.
- `Dense(4, input_dim=2, activation='relu')`: This adds a dense layer with
		 4 neurons, expecting input of dimension 2 (specified by `input_dim`).
		 The activation function used here is Rectified Linear Unit (ReLU).
- `Dense(1, activation='sigmoid')`: This adds the output layer with 1 neuron
		 and a sigmoid activation function. Sigmoid is often used in binary
		 classification problems as it squashes the output between 0 and 1.
'''
model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # This is the last layer, it should have 1 neuron and the sigmoid activation

# Step 2: Compile the model (Parameter Setting)
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.05), # Algorithm used for optimization
    loss='binary_crossentropy',                         # The loss function
    metrics=['accuracy'],                               # Metrics to evaluate the goodness of predictions
)

# Step 3: Train the Model
history = model.fit(
    Xtr, Ytr,                       # Training data
    epochs=600,                     # Number of training epochs (period)
																			# Number of steps in gradient descent 
    batch_size=10,                  # Train using mini-batches of 10 samples in each iteration
    validation_split=0.2,           # Split the data using 80% to train and 20% for validation
)

# Step 4: Predict and calculate errors
train_preds = model.predict(Xtr)
test_preds = model.predict(Xte)
train_err = calc_err(train_preds, Ytr)
test_err = calc_err(test_preds, Yte)

# Step 5: Print the error
print("Training error: %.2f%%, Test error: %.2f%%" % (train_err * 100, test_err * 100))
```

```python
# With this function you can plot the history of the model
	# training produced by the fit function
def plot_history(history):
    fig, ax = plt.subplots()
    # Plot training & validation accuracy values
    ax.plot(history.history['accuracy'], label='Train')
    ax.plot(history.history['val_accuracy'], label='Val')
    ax.set_title('Model accuracy')
    ax.set_ylabel('Accuracy')
    ax.set_xlabel('Epoch')
    ax.legend(loc='best')
    return fig, ax
```

```python
fig, ax = separatingFLR(Xtr, Ytr, train_preds, model)
ax.set_title("Predicted class seperation");
fig, ax = plot_history(history)
ax.set_title("Accuracy evolution over epochs");
```

### Step 5: Train a Neural Network on MNIST!

Code: Part 1 - Load The Data

```python
# The data, split between train and test sets
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
print("Training set shapes: ", x_train.shape, y_train.shape)
print("Test set shapes: ", x_test.shape, y_test.shape)
print('x_train: {0}'.format(x_train))
print('y_train: {0}'.format(y_train))
print('x_test: {0}'.format(x_test))
print('y_test: {0}'.format(y_test))
```

Code: Part 2 - Bring the data into shape

```python
num_1 = 1
num_2 = 7

# Here we take only two digits from MNIST. 
# We will reduce the problem to binary classification since we only require two options not more.
Xtr = x_train[(y_train == num_1) | (y_train == num_2)]
Ytr = y_train[(y_train == num_1) | (y_train == num_2)]
Xts = x_test[(y_test == num_1) | (y_test == num_2)]
Yts = y_test[(y_test == num_1) | (y_test == num_2)]

# Reshape the data correctly
Xtr = Xtr.reshape(-1, 28*28)
Xts = Xts.reshape(-1, 28*28)

# Now Ytr depends on the value of the num_1 and num_2 are set to zero or one
	# We are substituting y values 
Ytr[Ytr == num_1] = 0
Ytr[Ytr == num_2] = 1
Yts[Yts == num_1] = 0
Yts[Yts == num_2] = 1

print('Xtr: {0}'.format(Xtr))
print('Ytr: {0}'.format(Ytr))
print('Xts: {0}'.format(Xts))
print('Yts: {0}'.format(Yts))
```

Code: Part 3 - Visualization

```python
fig, ax = plt.subplots()
'''
**Axes.imshow(*X*, *cmap=None*, *norm=None*, ***, *aspect=None*, *interpolation=None*, *alpha=None*, *vmin=None*, *vmax=None*, *origin=None*, *extent=None*, *interpolation_stage=None*, *filternorm=True*, *filterrad=4.0*, *resample=None*, *url=None*, *data=None*, ***kwargs*)[[source]](https://github.com/matplotlib/matplotlib/blob/v3.8.2/lib/matplotlib/axes/_axes.py#L5537-L5769)**
Display data as an image, i.e., on a 2D regular raster.
'''
ax.imshow(Xtr[3].reshape(28, 28))
ax.set_title(Ytr[3]);
```

Code: Part 4 - Define The Keras Model

```python
# Create a Sequential model
model = Sequential()

# Add the first hidden layer with 100 neurons and ReLU activation
model.add(Dense(100, input_dim=28*28, activation='relu'))

# Add additional hidden layers with 30 neurons and ReLU activation
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))
# Add more hidden layers if needed

# This is the last layer, it should have 1 neuron and the sigmoid activation
model.add(Dense(1, activation='sigmoid'))
```

Code: Part 5: Compile The Keras Model

```python
model.compile(
    loss='binary_crossentropy', 
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
    metrics=['accuracy'],)
```

Code: Part 6 - Fit the model

```python
# fit the keras model on the dataset
history = model.fit(Xtr, Ytr, epochs=500, batch_size=10, validation_split=0.2)
```

Code: Part 7 - Prediction and Evaluation

```python
# Predict and calculate errors (if needed)
train_preds = model.predict(Xtr)
test_preds = model.predict(Xts)
train_err = calc_err(train_preds, Ytr)
test_err = calc_err(test_preds, Yts)
print("Training error: %.2f%%, Test error: %.2f%%" % (train_err * 100, test_err * 100))

# evaluate the keras model
_, accuracy = model.evaluate(Xts, Yts)
print('Test accuracy: %.2f%%' % (accuracy * 100))
```

Code: Part 8 - Plot training history

```python
# Plot training history
plot_history(history)
```
