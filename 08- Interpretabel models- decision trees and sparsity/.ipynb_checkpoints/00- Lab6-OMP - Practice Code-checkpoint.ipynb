### Step 1 - Data generation

To generate a synthetic dataset suitable for use with variable selection, we first generate a subset of "relevant" features, and then concatenate with a second set of "dummy", irrelevant featuers. To this purpose,  we proceed as follows:
- Generate a dataset with the `mixGauss` function: use two Gaussians which are *close* with each other (see example below). Start by considering 2-Dimensional points
- Plot the points: you should observe two point-clouds which mix with each other (since they were generated with the same parameters).
- Enrich the input samples with "dummy" features randomly sampled and concatenate to the relevant features (notice that at this point data visualization is no more possible


```python
import numpy as np

def mixGauss(means, sigmas, n):
    # Convert means and sigmas to NumPy arrays
    means = np.array(means)
    sigmas = np.array(sigmas)

    # Get the dimensionality of the data -> columns
    # shape[0] = rows number
    # shape[1] = columns number
    # One of the lines is extra, one is sufficient
    d = means.shape[1]

    # Determine the number of classes
    num_classes = sigmas.size

    # Initialize arrays to store synthetic data and labels
    # np.full((2, 2), np.inf)
    #	array([[inf, inf],
    #	       [inf, inf]])
    data = np.full((n * num_classes, d), np.inf)
    labels = np.zeros(n * num_classes) # labels=y -> y contains one column

    # Generate synthetic data for each class
		# enumerate [1,2,3] -> [[0,1],[1,2],[2,3]]
    for idx, sigma in enumerate(sigmas):
    
        ''' 
        data[idx * n:(idx + 1) * n] = np.random.multivariate_normal(
        mean=means[idx], cov=np.eye(d) * sigmas[idx] ** 2, size=n)
        random.multivariate_normal(mean, cov, size=None, check_valid='warn', tol=1e-8)
        Draw random samples from a multivariate normal distribution.
        The multivariate normal, multinormal or Gaussian distribution is a generalization of the one-dimensional normal distribution to higher dimensions. Such a distribution is specified by its mean and covariance matrix. These parameters are analogous to the mean (average or “center”) and variance (standard deviation, or “width,” squared) of the one-dimensional normal distribution.
        '''
    
        ''' 
        --- np.random.multivariate_normal: This NumPy function generates random samples from a multivariate (چند متغییره) normal distribution.
        --- mean=means[idx]: The mean vector for the current class is specified by the 'means' array at index `idx`.
        --- cov=np.eye(d) * sigmas[idx] ** 2: The covariance matrix is a diagonal matrix with diagonal elements equal to the square of the current class's standard deviation (`sigma`). 
        --- np.eye(d)`:creates an identity matrix of size `d`.
        --- size=n: This parameter specifies the number of samples to generate for the current class.
        '''
    
        ''' numpy.eye(N, M=None, k=0, dtype=<class 'float'>, order='C', *, like=None)[source]
            Return a 2-D array with ones on the diagonal and zeros elsewhere.
            np.eye(2, dtype=int)
            array([[1, 0],
                   [0, 1]])
            np.eye(3, k=1)
            array([[0.,  1.,  0.],
                   [0.,  0.,  1.],
                   [0.,  0.,  0.]]) '''
            
        # random.multivariate_normal(mean, cov, size=None, check_valid='warn', tol=1e-8)
        # Draw random samples from a multivariate normal distribution.
        ''' Let's clarify this code with real example:
                    sigma_size = 2 (columns or classes)
                    idx = 0
                    data[0:100] = ...
                    label[0:100] = ...
    
                    idx = 1
                    data[100:200] = ...
                    label[100:200] = ...
                        as it is obvious from this generation is accomplished for each class first 0 to 100 elements of the array, the second is 100 to 200. 
                    Actually, instead of using nested for, we use range to assign the random values
        '''
        # Generate n samples for each class using multivariate normal distribution
        data[idx * n:(idx + 1) * n] = np.random.multivariate_normal(
            mean=means[idx], cov=np.eye(d) * sigmas[idx] ** 2, size=n)
        
        # Assign class labels -> [n:n+1] means just nth row 
        labels[idx * n:(idx + 1) * n] = idx 

    # If there are only two classes, adjust labels to -1 for one class
    if num_classes == 2:
				# The labels (y) which are zeros in the comprehensive array (labels)
					# equal to -1 -> actually we changed the value from 0 to -1
        labels[labels == 0] = -1

    # Return the generated data and labels
    return data, labels
```

```python
n=1000 # number of features for each gaussian
d=30 # total number of features
d_rev = 2 # number of relevant features

Xtr, Ytr = mixGauss(means = [[0,0],[2,2]], sigmas = [0.9, 0.75], n=n)
Xte, Yte = mixGauss(means = [[0,0],[2,2]], sigmas = [0.9, 0.75], n=n)

# plot
fig, ax = plt.subplots()
ax.set_title("Training Set", fontsize=22)
ax.scatter(Xtr[Ytr==-1,0], Xtr[Ytr==-1,1], s=30, alpha=0.80)
ax.scatter(Xtr[Ytr==1,0], Xtr[Ytr==1,1], s=30, alpha=0.80)


fig, ax = plt.subplots()
ax.set_title("Test Set", fontsize=22)
ax.scatter(Xte[Yte==-1,0], Xte[Yte==-1,1], s=30, alpha=0.80)
ax.scatter(Xte[Yte==1,0], Xte[Yte==1,1], s=30, alpha=0.80)
```

## Variable selection

##### OMatchingPursuit

It computes a sparse representation of the signal using Orthogonal Matching Pursuit algorithm. Use it as follows:
    
`w, r, I = OMatchingPursuit( X, Y, T)`

where
    - X: input data
    - Y: output labels
    - T: number of iterations
    - w: estimated coefficients
    - r: residuals
    - I: indices


'''
In summary, the residual measures the unexplained part of the target variable
, while the coefficient vector represents the weights assigned to each feature
in the model. The goal of algorithms like Orthogonal Matching Pursuit is to 
iteratively select features and update coefficients to minimize the residual, 
ultimately capturing the underlying patterns in the data.
'''
def OMatchingPursuit(X, Y, T):
		
    # Step 1: Initialization	
    # N = rows 
    # D = columns
    N, D = np.shape(X)

    # 1. Initialization of residual, coefficient vector, and index set I
    '''
        1. **Residual:**
        - The residual is the difference between the predicted values and the actual values in a regression problem.
        - In the context of the Orthogonal Matching Pursuit algorithm, the residual (\(r\)) represents the part of the target variable (\(Y\)) that has not been explained by the selected features and their corresponding coefficients.
    '''
    r = Y
    w = np.zeros(D)
    I = [] # I is selected features
    
    # Step 2: Feature Selection
    for i in range(T):
        '''
        Sets are used to store multiple items in a single variable.
        # Unordered, Unchangeable, 
        myset = {"apple", "banana", "cherry"}
        '''
        I_tmp = list(set(range(D)) - set(I))  # Exclude the indices already selected

        # 2. Select the column of X which coefficients most "explains" the residual
				# Embark on -1
        a_max = -1
        j_max = None

        for j in I_tmp:
            # Calculate the inner product of the j-th column of X with the residual
            ''' 
                X[:, j].T @ r performs the matrix multiplication of the transposed j-th column of X with the vector r.
                np.linalg.norm calculates the Euclidean norm of a vector or matrix.
            '''
            a_tmp = np.abs(X[:, j].T @ r) / np.linalg.norm(X[:, j])

            if a_tmp > a_max:
                a_max = a_tmp
                j_max = j

        # 3. Add the index to the set of indexes (the index is selected)
        if j_max is not None:
            I.append(j_max)
				
        # Step 3- Matrix Operations
        # Compute the M matrix
        # - The matrix **`M_I`** is created, representing the selection
         # of features in matrix form.
        M_I = np.zeros((D, D))
				
        # M_I is a matrix with ones on the diagonal for the selected features
        for j in I:
            M_I[j, j] = 1
				
            # Matrices A and B are computed to update the coefficients.
            '''
                5. **`M_I.dot(X.T).dot(Y)`:**
        - Similarly, this line computes the dot product of the matrix `M_I` with
                 the transpose of `X` and then with the target variable `Y`.
            '''
            # Aw = B
            # dot(M_I) is 1 only on diagnol part -> the multiplication makes other values
            # to zeros
        A = M_I.dot(X.T).dot(X).dot(M_I)
        B = M_I.dot(X.T).dot(Y)

        # 4. Update estimated coefficients
        # The estimated coefficients (w) are updated by solving the linear system A * w = B.
        w = np.linalg.solve(A, B)

        # 5. Update the residual
        '''
             The formula calculates the difference between the actual target values (\(Y\)) and the predicted values (\(X \cdot w\))
        '''
        r = Y - X.dot(w)

    return w, r, I

```python
# ChatGPT Code 
# ChatGPT Code
def OMatchingPursuit(X, Y, T, alpha=1e-5):
    N, D = np.shape(X)

    # 1. Initialization of residual, coefficient vector, and index set I
    r = Y
    w = np.zeros(D)
    I = []

    for i in range(T):
        I_tmp = list(set(range(D)) - set(I))

        # 2. Select the column of X which coefficients most "explains" the residual
        a_max = -1
        j_max = -1

        for j in I_tmp:
            a_tmp = np.abs(X[:, j].T @ r) / np.linalg.norm(X[:, j]) ** 2

            if a_tmp > a_max:
                a_max = a_tmp
                j_max = j

        # 3. Add the index to the set of indexes
        if j_max not in I:
            I.append(j_max)

        # Compute the M matrix
        M_I = np.zeros((D, D))

        for j in I:
            M_I[j, j] = 1

        A = M_I.dot(X.T).dot(X).dot(M_I) + alpha * np.eye(D)
        B = M_I.dot(X.T).dot(Y)

        # 4. Update estimated coefficients with Ridge regularization
        w = np.linalg.solve(A, B)

        # 5. Update the residual
        r = Y - X.dot(w)

    return w, r, I
```

```python
# Usual function to compute the error in predicted labels (for a binary classification problem)
def calcError(Ypred, Y):
    V = np.multiply(np.sign(Ypred), np.sign(Y))
    return np.count_nonzero(V < 0) / len(Y)
```

```python
# K-Fold Cross Validation for selecting the best value for the hyperparameters in the use of 
# Orthogonal Matching Pursuit (we refer to the number of iterations)

def KFoldCVOMP(Xtr, Ytr, KF, niter_list):
    if KF <= 0:
        print("Please supply a positive number of repetitions")
        return -1

    # Ensures that k_list is a numpy array
    niter_list = np.array(niter_list)
    num_niter = niter_list.size

    n_tot = Xtr.shape[0]
    n_val = int(np.floor(n_tot/KF)) # For 200 elements and kf = 5 -> 200/5 = 40 indecies for each fold

    Tm = np.zeros(num_niter)
    Ts = np.zeros(num_niter)
    Vm = np.zeros(num_niter)
    Vs = np.zeros(num_niter)

    # Random permutation of training data
    '''		
    >>> import numpy as np
    >>> rnd_state = np.random.RandomState(12)
    >>> rand_idx = rnd_state.choice(10, size=10, replace=False)
    >>> rand_idx
    array([5, 8, 7, 0, 4, 9, 3, 2, 1, 6])
    '''
    # Random n_tot index generation by n_tot size for all the folds (groups)
    rand_idx = rnd_state.choice(n_tot, size=n_tot, replace=False)
	   
    # enumerate is utilized to set index for each value 
    # [1,5,7] -> [[0,1],[1,5],[2,7]] 
    for kdx, niter in enumerate(niter_list):
        first = 0
        for fold in range(KF):
            flags = np.zeros(Xtr.shape[0])
            # first:first+n_val -> each fold for instance first = 0 and each fold with 40 elements that would be: [0, 40]
            # Now with this code (flag) current fold or group flag array is set to one and others are zero (already it was initilized with zeros)
            flags[first:first+n_val]=1;
        
            # Fro rand_idx array for Xtr choose the elements that the flags values equal to 0 (current fold) -> the major section is assigned for learning and the left part is assigned to testing
            X = Xtr[rand_idx[flags==0]]
            Y = Ytr[rand_idx[flags==0]]
            X_val = Xtr[rand_idx[flags==1]]
            Y_val = Ytr[rand_idx[flags==1]]

            # Compute the training error of OMP for the given number of iterations
            w, r, I = OMatchingPursuit(X, Y, niter)
            YpredTR = np.sign(X.dot(w))
            trError = calcError(YpredTR, Y)
            Tm[kdx] = Tm[kdx] + trError
            Ts[kdx] = Ts[kdx] + trError ** 2

            # Compute the validation error OMP for the given number of iterations
            YpredVAL = np.sign(X_val.dot(w))
            valError = calcError(YpredVAL, Y_val)
            Vm[kdx] = Vm[kdx] + valError
            Vs[kdx] = Vs[kdx] + valError ** 2
            
            # To embark on the first element of the next fold -> for instance the first one was [0, 39] now moves to [40,79]
            first = first+n_val                

    Tm = Tm / KF
    Ts = Ts / KF - Tm ** 2

    Vm = Vm / KF
    Vs = Vs / KF - Vm ** 2

    best_niter_idx = np.argmin(Vm)
    bestniter = niter_list[best_niter_idx]
    
    print(Vm)
    print(niter_list)
    print(best_niter_idx)

    return bestniter, Vm, Vs, Tm, Ts
```

### Some analysis

We suggest to proceed as follows:

- Standardize the data
- Run Orthogonal Matching Pursuit on the training set setting a reasonable number of iterations
- Compute the prediction on the test set and evaluate the error
- Plot the components of the solution w (considering their absolute value): how do they look?
- Run the K-Fold Cross Validation to select an appropriate value for the number of iterations


```python
# Data standardization
m = np.mean(Xtr, axis=0)
s = np.std(Xtr, axis=0)
Xtrnorm = np.divide((Xtr - m), s)
Xtenorm = np.divide((Xte - m), s)

# Run Orthogonal Matching Pursuit
n_iterations_omp = 5  # Set a reasonable number of iterations
w_omp, r_omp, I_omp = OMatchingPursuit(Xtrnorm, Ytr, n_iterations_omp)

# Compute the prediction on the test set
Ypred_omp = np.sign(Xtenorm.dot(w_omp))

# Compute the test error and show its value
test_error_omp = calcError(Ypred_omp, Yte)
print(f"Test error with OMP: {test_error_omp}")
```

```python
# Run K-Fold Cross Validation
KF = 5
niter_list = [1, 2, 3, 4, 5, 10, d]
bestniter, Vm, Vs, Tm, Ts = KFoldCVOMP(Xtrnorm, Ytr, KF, niter_list)

# Plot training and validation error
plt.figure(figsize=(10, 6))
# For training -> x = niter_list (number of iterations), y = Tm (Error rate in each iteration) 
plt.plot(niter_list, Tm, label="training")
# For validation -> x = niter_list (number of iterations), y = Tm (Error rate in each iteration) 
plt.plot(niter_list, Vm, label="validatitrainingon")
plt.legend(loc="best")
plt.xlabel("Number of Iterations")
plt.ylabel("Error Rate")
# An simple vertical line as the best iteration for error calculation
plt.axvline(bestniter, color="red", linestyle="--", label="Best Iteration")
plt.title("Training and Validation Error Rates vs. Number of Iterations")
plt.show()
```
